##Descripccion de los 10 materiales con mayor extracción
* **Steam Coal (carbón térmico)**: El carbón térmico (a veces llamado carbón vapor) es un tipo de carbón que se utiliza en las centrales eléctricas para generar vapor y generar electricidad. La mayoría de los carbones extraídos en los Estados Unidos son carbones vapor. Los carbones vapor para las centrales eléctricas deben cumplir con las características de calidad y calefacción del diseño de la caldera y del equipo de control de la contaminación en una central eléctrica.

* **Petroleum (Petroleo)**: El petróleo (del griego: πετρέλαιον, lit. «aceite de roca») es una mezcla de compuestos orgánicos, principalmente hidrocarburos insolubles en agua. se produce en el interior de la Tierra, por transformación de la materia orgánica acumulada en sedimentos del pasado geológico1​ y puede acumularse en trampas geológicas naturales, que se localizan a nivel mundial2​ y de donde se extrae mediante la perforación de pozos.

* **Iron(hierro)**: Es un metal maleable, de color gris plateado, y presenta propiedades magnéticas (es ferromagnético a temperatura ambiente y presión atmosférica). Es extremadamente duro y denso. Se encuentra en la naturaleza formando parte de numerosos minerales, entre ellos muchos óxidos, y raramente se encuentra libre. Para obtener hierro en estado elemental, los óxidos se reducen con carbono y luego es sometido a un proceso de refinado para eliminar las impurezas presentes.

* **Coking Coal (carbón de coque)**: es un tipo de carbón que se puede utilizar para producir coque de buena calidad . El coque es un combustible y reactivo esencial en el proceso de alto horno para la fabricación de acero primario. La demanda de carbón metalúrgico está muy relacionada con la demanda de acero. Las empresas de fabricación de acero primario suelen tener una división que produce carbón para coquización, para garantizar un suministro estable y de bajo costo.

* **Lignite(Lignite)**: El lignito (derivado del latín lignum que significa 'madera'), a menudo llamado carbón pardo, es una roca sedimentaria blanda, marrón y combustible formada a partir de turba comprimida naturalmente . Tiene un contenido de carbono de alrededor del 25-35% y se considera el rango más bajo de carbón debido a su contenido de calor relativamente bajo . Cuando se extrae del suelo, contiene una cantidad muy alta de humedad , lo que explica parcialmente su bajo contenido de carbono. El lignito se extrae en todo el mundo y se utiliza casi exclusivamente como combustible para la generación de energía eléctrica a vapor.

* **Bauxite(bauxita)**: La bauxita es una roca sedimentaria con un contenido de aluminio relativamente alto. Es la principal fuente de aluminio del mundo. La bauxita consiste principalmente en los minerales de aluminio gibbsita (Al(OH)3), boehmita (γ-AlO(OH)) y diásporo (α-AlO(OH)), mezclados con dos óxidos de hierro, goetita y hematita, minerales de la arcilla de aluminio, caolinita, y pequeñas cantidades de anatasa (TiO2 ) e ilmenita (FeTiO3 o FeO.TiO2).

* **Salt(sal)**: La sal es la única roca que es comestible para el ser humano, y es posiblemente el condimento más antiguo. Su importancia para la vida es tal que ha marcado el desarrollo de la historia en muchas ocasiones, y sigue moviendo las economías y es objeto de impuestos, monopolios, guerras, etcétera Fue incluso un tipo de moneda. El valor que tuvo en la antigüedad se redujo desde que comenzó a disminuir su demanda mundial para el consumo humano, debido en parte a la mejora en su producción y a la conciencia mundial que ha generado la posible relación que posee con la aparición de la hipertensión.9​En el siglo XXI, las dietas procuran incluir menos sal en sus composiciones, y los nuevos sistemas de conservación (pasteurizados, refrigerados y congelados, alimentos envasados al vacío, etcétera) permiten evitar por completo el empleo de la salazón sobre los alimentos.


* **Oil Sands(arenas petrolíferas)**: Las arenas petrolíferas [ a ] ​​son ​​un tipo deyacimiento de petróleo no convencional. Son arenas sueltas oareniscasparcialmente consolidadas que contienen una mezcla natural dearena,arcillayagua, empanadas conbetún(una forma densa y extremadamenteviscosa de petróleo).

* **Gypsum(yeso)**: El yeso, piedra de yeso, yeso crudo, yeso natural o aljez es un mineral compuesto de sulfato de calcio hidratado; también, una roca sedimentaria de origen químico. Es un mineral muy común y puede formar rocas sedimentarias monominerales. El yeso mineral cristaliza en el sistema monoclínico, en cristales de hábito prismático; tabular paralelo al segundo pinacoide; de forma rómbica con aristas biseladas en las caras.4​ Se presenta en cristales, a veces grandes, maclados en punta de flecha y en punta de lanza; también en masas y agregados espáticos. Con frecuencia fácilmente exfoliable (selenita); puede ser sacaroideo y translúcido (alabastro), incoloro, blanco, o de distintos colores, grisáceo, amarillento, rojizo o incluso negro, dependiendo de las inclusiones que contenga.

* **Diam. (Gem)(diamante)**: En mineralogía, el diamante es un mineral alótropo del carbono en el que los átomos de carbono están dispuestos en una variante de la estructura cristalina cúbica centrada en la cara denominada. El diamante es la segunda forma más estable de carbono, después del grafito; sin embargo, la tasa de conversión de diamante a grafito es despreciable a condiciones ambientales. El diamante tiene renombre específicamente como un material con características físicas superlativas, muchas de las cuales derivan del fuerte enlace covalente entre sus átomos. En particular, el diamante tiene la más alta dureza y conductividad térmica de todos los materiales conocidos por el ser humano. Estas propiedades determinan que la principal aplicación industrial del diamante sea en herramientas de corte y de pulido, además de otras aplicaciones. Un diamante de calidad superior es denominado un naife.

##Grafico del top 5 países con mayor extracción

```python
from matplotlib import pyplot as plt
import seaborn as sns

def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['year']
  ys = series['extraction_amount']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

material_raw = top_10_materiales.index.get_level_values('mined_raw_mat').tolist()

for material in material_raw:

  df_material = df[df['mined_raw_mat'] == material]

  top_10_paises = df_material.groupby('country')['Total_extracccion'].sum().sort_values(ascending=False).head(5)

  df_top_10_paises = df[df['country'].isin(top_10_paises.index) & (df['mined_raw_mat'] == material)]

  df_material_derretido = pd.melt(df_top_10_paises,
                                  id_vars=['country', 'mined_raw_mat', 'unit', 'Total_extracccion'],
                                  value_vars=['2018', '2019', '2020', '2021', '2022'],
                                  var_name='year',
                                  value_name='extraction_amount')
  fig, ax = plt.subplots(figsize=(8, 4), layout='constrained')
  df_sorted = df_material_derretido.sort_values('year', ascending=True)
  for i, (series_name, series) in enumerate(df_sorted.groupby('country')):
    _plot_series(series, series_name, i)

  fig.legend(title='country', bbox_to_anchor=(1, 1), loc='upper left')
  sns.despine(fig=fig, ax=ax)
  plt.xlabel('Año')
  _ = plt.ylabel('Cantidad de Extraccion')

  plt.title(f'Top 5 países con mayor extracción de {material}')

  plt.tight_layout()
  plt.show()

```

```python
import pandas as pd

df_empresa['Commodity'] = df_empresa['Commodity'].astype(str)

materiales_explotados = df_empresa['Commodity'].str.split(',').explode().str.strip().value_counts()

top_5_materiales = materiales_explotados.head(5)

top_5_materiales


```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_empresa['Commodity'] = df_empresa['Commodity'].astype(str)


materiales_explotados = df_empresa['Commodity'].str.split(',').explode().str.strip().value_counts()


top_5_materiales = materiales_explotados.head(5)


empresas_top_5_materiales = df_empresa[df_empresa['Commodity'].apply(
    lambda x: any(material in [material_loc.strip() for material_loc in x.split(',')] for material in top_5_materiales.index)
)]

empresas_top_5_materiales_df = empresas_top_5_materiales[['Name', 'Location', 'Commodity']].drop_duplicates()

empresas_top_5_materiales_df['Location'] = empresas_top_5_materiales_df['Location'].str.split(',')


empresas_por_material = {}
for material in top_5_materiales.index:
    empresas_material = empresas_top_5_materiales_df[empresas_top_5_materiales_df['Commodity'].apply(
        lambda x: material in [material_loc.strip() for material_loc in x.split(',')]
    )]
    empresas_por_material[material] = empresas_material['Name'].nunique()


print("Cantidad de empresas distintas que extraen los 5 materiales más extraídos:")
for material, num_empresas in empresas_por_material.items():
    print(f"{material}: {num_empresas} empresas")


plt.figure(figsize=(10, 6))
plt.bar(empresas_por_material.keys(), empresas_por_material.values(), color='lightcoral')

plt.title('Cantidad de empresas distintas que extraen los 5 materiales más extraídos', fontsize=14)
plt.xlabel('Material', fontsize=12)
plt.ylabel('Cantidad de empresas distintas', fontsize=12)
plt.xticks(rotation=45, ha='right')


plt.tight_layout()
plt.show()
```

##Descripccion de los materiales

* **Oro**: El oro es uno de los metales más densos. Es un buen conductor del calor y la electricidad. También es blando y el más maleable y dúctil de los elementos; una onza (31,1 gramos; el oro se pesa en onzas troy ) puede convertirse en 187 pies cuadrados (unos 17 metros cuadrados) en láminas extremadamente finas llamadas pan de oro .

* **Cobre**: El cobre nativo se encuentra en muchos lugares como mineral primario en lavas basálticas y también como reducido a partir de compuestos de cobre , como sulfuros, arseniuros, cloruros y carbonatos. (Para las propiedades mineralógicas del cobre, consulte la tabla de elementos nativos ). El cobre se presenta combinado en muchos minerales , como calcocita , calcopirita , bornita , cuprita , malaquita y azurita . Está presente en las cenizas de las algas marinas , en muchos corales marinos, en el hígado humano y en muchos moluscos y artrópodos . El cobre desempeña el mismo papel de transporte de oxígeno en la hemocianina de los moluscos y crustáceos de sangre azul que el hierro en la hemoglobina de los animales de sangre roja. El cobre presente en los humanos como oligoelemento ayuda a catalizar la formación de hemoglobina. Un depósito de pórfido de cobre en la Cordillera de los Andes de Chile es el mayor depósito conocido del mineral . A principios del siglo XXI, Chile se había convertido en el principal productor mundial de cobre. Otros productores importantes eran Perú, China y Estados Unidos .

* **Plata**: Junto con el oro y los metales del grupo del platino , la plata es uno de los llamados metales preciosos . Debido a su relativa escasez, su color blanco brillante, su maleabilidad, ductilidad y resistencia a la oxidación atmosférica, la plata se ha utilizado durante mucho tiempo en la fabricación de monedas , adornos y joyas . La plata tiene la conductividad eléctrica y térmica más alta conocida de todos los metales y se utiliza en la fabricación de circuitos eléctricos impresos y como revestimiento depositado por vapor para conductores electrónicos ; también se alea con elementos como el níquel o el paladio para su uso en contactos eléctricos. La plata también se utiliza como catalizador por su capacidad única de convertir el etileno en óxido de etileno, que es un precursor de muchos compuestos orgánicos . La plata es uno de los elementos de transición más nobles, es decir, menos reactivos químicamente.

* **Nickel**: De color blanco plateado, resistente y más duro que el hierro , el níquel es muy conocido por su uso en la acuñación de monedas, pero es más importante ya sea como metal puro o en forma de aleaciones por sus numerosas aplicaciones domésticas e industriales. El níquel elemental se encuentra muy escasamente junto con el hierro en depósitos terrestres y meteóricos. El metal fue aislado (1751) por un químico y mineralogista sueco, el barónAxel Fredrik Cronstedt , quien preparó una muestra impura de un mineral que contenía niccolita (arseniuro de níquel). Anteriormente, un mineral de este mismo tipo se llamaba Kupfernickel en honor al “viejo Nick” y sus traviesos gnomos porque, aunque se parecía al mineral de cobre, producía un metal frágil y desconocido. El níquel, que es dos veces más abundante que el cobre, constituye aproximadamente el 0,007 por ciento de la corteza terrestre; es un componente bastante común de las rocas ígneas, aunque singularmente pocos depósitos califican en concentración, tamaño y accesibilidad para el interés comercial. Se cree que las regiones centrales de la Tierra contienen cantidades considerables. Las fuentes más importantes sonpentlandita , que se encuentra con pirrotita que contiene níquel , de la cual ciertas variedades contienen entre un 3 y un 5 por ciento de níquel, y calcopirita , y lateritas que contienen níquel, como la garnierita, un silicato de magnesio y níquel de composición variable.

* **Litio**: El litio es un metal clave para la transición energética a nivel global, en particular para la electromovilidad y el almacenamiento de energía obtenida a partir de fuentes renovables. Su importancia se acrecentó cuando comenzó a destinarse a la producción de baterías para equipos electrónicos portables y, en forma más reciente, cuando se convirtió en un insumo esencial para la fabricación de las baterías utilizadas en los vehículos eléctricos.
